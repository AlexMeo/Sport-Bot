```python
@bot.message_handler(commands=['start'])
def start(message):
    # keyboard
    markup = types.ReplyKeyboardMarkup(resize_keyboard=True)
    item1 = types.KeyboardButton("–§—É—Ç–±–æ–ª‚öΩ")
    item2 = types.KeyboardButton("–¢–µ–Ω–Ω–∏—Åüéæ")
    item3 = types.KeyboardButton("–•–æ–∫–∫–µ–πüèí")
    
    #markup.add(item1, item2, item3, item4)
    markup.add(item1).add(item2).add(item3)

    bot.send_message(message.chat.id, "–î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å, {0.first_name}!\n–Ø - <b>{1.first_name}</b>, —á—Ç–æ –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç?".format(message.from_user, bot.get_me()),
        parse_mode='html', reply_markup=markup)
```
