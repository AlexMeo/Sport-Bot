```python
import telebot
import config
import pandas as pd
import numpy as np
#from datetime import datetime
#from time import sleep
import random
from telebot import types
```

```python
bot = telebot.TeleBot(config.TOKEN)
data = pd.read_excel("C:\Databaza.xlsx")
data.head
```

```python
@bot.message_handler(commands=['start'])
def start(message):
    # keyboard
    markup = types.ReplyKeyboardMarkup(resize_keyboard=True)
    item1 = types.KeyboardButton("Футбол⚽")
    item2 = types.KeyboardButton("Теннис🎾")
    item3 = types.KeyboardButton("Хоккей🏒")
    
    #markup.add(item1, item2, item3, item4)
    markup.add(item1).add(item2).add(item3)

    bot.send_message(message.chat.id, "Добро пожаловать, {0.first_name}!\nЯ - <b>{1.first_name}</b>, что интересует?".format(message.from_user, bot.get_me()),
        parse_mode='html', reply_markup=markup)
```

```python
@bot.message_handler(content_types=['text'])
def message(message):
    if message.chat.type == 'private':

        if message.text == 'Футбол⚽':
            #Кливиатура футбольных чемпионатов исчезающая после нажатия
            markup = types.ReplyKeyboardMarkup(resize_keyboard=True, one_time_keyboard=True)
            item1 = types.KeyboardButton("Россия")
            item2 = types.KeyboardButton('Англия')
            item3 = types.KeyboardButton("Италия")
            item4 = types.KeyboardButton("Испания")
            item5 = types.KeyboardButton("Германия")
            item6 = types.KeyboardButton("/start")
            markup.add(item1)
            markup.row(item2, item3)
            markup.row(item4, item5)
            markup.row(item6)
            bot.send_message(message.chat.id, 'Отлично, какой чемпионат?', reply_markup=markup)
            
        elif message.text == 'Теннис🎾':
            #Кливиатура турниров большого шлема исчезающая после нажатия
            markup = types.ReplyKeyboardMarkup(resize_keyboard=True, one_time_keyboard=True)
            item1 = types.KeyboardButton("Australian Open🦘")
            item2 = types.KeyboardButton("Rolland Garros🐸")
            item3 = types.KeyboardButton("Wimbledon🦁")
            item4 = types.KeyboardButton("US Open🦅")
            item5 = types.KeyboardButton("/start")
            markup.add(item1, item2)
            markup.row(item3, item4)
            markup.row(item5)
            bot.send_message(message.chat.id, 'Респект, какой турнир?', reply_markup=markup)
            
        #elif message.text == 'Баскетбол🏀':
            #bot.send_message(message.chat.id, 'нет баскетбола')
            
        elif message.text == "Хоккей🏒":
            #Кливиатура хоккейных чемпионатов исчезающая после нажатия
            markup = types.ReplyKeyboardMarkup(resize_keyboard=True, one_time_keyboard=True)
            item1 = types.KeyboardButton("Кубок Стэнли")
            item2 = types.KeyboardButton("Кубок Гагарина")
            markup.add(item1, item2)
            bot.send_message(message.chat.id, 'Где будем смотреть чемпионов?', reply_markup=markup)
            
        elif (message.text == 'Англия') or (message.text =="Италия") or (message.text =='Испания') or (message.text =='Германия') :
            global a
            bot.send_message(message.chat.id, 'Не лучший выбор, но ладно, какой год?')
            if message.text == 'Англия':
                a=1
            elif message.text == "Испания":
                a=2
            elif message.text == "Германия":
                a=3
            elif message.text == "Италия":
                a=5
            
        elif message.text == 'Россия' :
            a = 4
            bot.send_message(message.chat.id, 'А вы гурман! Какой год?')
           
            
        elif (message.text == "Australian Open🦘") or (message.text =="Rolland Garros🐸") or (message.text =="Wimbledon🦁") or (message.text =="US Open🦅"):
            bot.send_message(message.chat.id, 'Супер, в каком году?')
            if message.text == "Australian Open🦘":
                a=6
            elif message.text == "Rolland Garros🐸":
                a=7
            elif message.text == "Wimbledon🦁":
                a=8
            elif message.text == "US Open🦅":
                a=9
                
        elif (message.text == 'Кубок Стэнли') or (message.text =="Кубок Гагарина"):
            if message.text == 'Кубок Стэнли':
                a=10
            elif message.text == "Кубок Гагарина":
                a=11
            bot.send_message(message.chat.id, 'Супер, в каком году?')
            
            #Границы по таблицы по времени и вывод нужной информации
        elif (int(message.text) >= 1992) and (int(message.text) <= 2022) :
            b = int(message.text)
            #bot.send_message(message.chat.id,  data.iloc[2022-b, 1])
            bot.send_message(message.chat.id, '🏆'+ data.iloc[2022-b, 2*a-1])
            bot.send_message(message.chat.id,  data.iloc[2022-b, 2*a])
            markup = types.ReplyKeyboardMarkup(resize_keyboard=True, one_time_keyboard=True)
            item5 = types.KeyboardButton("/start")
            markup.add(item5)
            bot.send_message(message.chat.id, 'Другой год или на старт?', reply_markup=markup)
            
        elif int(message.text) > 2022 :
            b = int(message.text)
            bot.send_message(message.chat.id, 'Достоверной информации нет, но вы можете обратиться к эксперту за прогнозом t.me/alexey_meo')
            
        elif int(message.text) < 1992 :
            b = int(message.text)
            bot.send_message(message.chat.id, 'ой, это давно было...')
            
        elif message.text == "В меню" :
            markup = types.ReplyKeyboardMarkup(resize_keyboard=True)
            item1 = types.KeyboardButton("Футбол⚽")
            item2 = types.KeyboardButton("Теннис🎾")
            item3 = types.KeyboardButton("Хоккей🏒")
    
            #markup.add(item1, item2, item3, item4)
            markup.add(item1).add(item2).add(item3)
            bot.send_message(message.chat.id, "Ок, что интересует?", reply_markup=markup)
            
        else:
            bot.send_message(message.chat.id, message.text)
```
